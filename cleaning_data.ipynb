# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("genome_train.csv")
test = pd.read_csv("genome_test.csv")
```

```python
train.info()
```

```python
test.info()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
(test == "-99").sum()
```

```python
test.replace("-99", np.nan, inplace=True)
```

```python
train.drop(columns=["Patient Id", "Institute Name", "Family Name"], inplace=True)
test.drop(columns=["Patient Id", "Institute Name", "Family Name"], inplace=True)
```

```python
train.columns
```

```python
num_cols = train.select_dtypes(include=["int64", "float64"]).columns
```

```python
for col in num_cols:
    print(col)
```

```python
train[num_cols] = train[num_cols].fillna(train[num_cols].median())
test[num_cols] = test[num_cols].fillna(train[num_cols].median())
```

```python
train.isnull().sum().sum()
```

```python
test.isnull().sum().sum()
```

```python
cat_cols = train.select_dtypes(include='str').columns
```

```python
print(cat_cols)
```

```python
for col in cat_cols:
    print(col)
```

```python
train.drop(columns=["Disorder Subclass"], inplace=True, errors="ignore")
test.drop(columns=["Disorder Subclass"], inplace=True, errors="ignore")
```

```python
cat_cols = train.select_dtypes(include="str").columns
```

```python
print(cat_cols)
```

```python
cat_cols = train.select_dtypes(include="object").columns.tolist()
```

```python
cat_cols.remove("Genetic Disorder")
```

```python
if "Disorder Subclass" in cat_cols:
    cat_cols.remove("Disorder Subclass")
```

```python
print(cat_cols)
```

```python
for col in cat_cols:
    train[col] = train[col].fillna(train[col].mode()[0])
    test[col] = test[col].fillna(train[col].mode()[0])
```

```python
train["Genetic Disorder"] = train["Genetic Disorder"].fillna(
    train["Genetic Disorder"].mode()[0]
)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train.to_csv("train_cleaned.csv", index=False)
test.to_csv("test_cleaned.csv", index=False)
```
